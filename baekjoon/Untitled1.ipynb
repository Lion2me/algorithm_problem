```python
from collections import deque, defaultdict

MAX_COUNT = 1000000000
def get_diff_rate(p1, p2):
    try:
        diff = [p2[0] - p1[0],p2[1] - p1[1]]
        diff_rate = [diff[0]/sum(diff), diff[1]/sum(diff)]
    except ZeroDivisionError as e:
        return -1
    return diff_rate

def solve():
    n = int(input())
    team1 = list(map(int, input().split(' ')))
    team2 = list(map(int, input().split(' ')))
    visited = [MAX_COUNT] * n
    p1s = []
    p2s = []
    answer = 0
    
    pass_path = defaultdict(list)
    
    for i in range(0,n):
        p1s.append([i,tuple(team1[i*2:(i+1)*2])])
        p2s.append([i+n,tuple(team2[i*2:(i+1)*2])])
    
    all_p = p1s + p2s
    
    for idx, p1 in p1s:
        for target_idx, target_p1 in p1s:
            if idx == target_idx:
                continue
            enable_pass = True
            max_x = max(p1[0],target_p1[0])
            max_y = max(p1[1], target_p1[1])
            min_x = min(p1[0],target_p1[0])
            min_y = min(p1[1], target_p1[1])
            diff_rate = get_diff_rate(p1,target_p1)
            for block_idx, block_p in all_p:
                if block_idx == target_idx or block_idx == idx:
                    continue
                    
                diff_rate_block = get_diff_rate(p1, block_p)
                if diff_rate_block == diff_rate and block_p[0] >= min_x and block_p[0] <= max_x and block_p[1] >= min_y and block_p[1] <= max_y:
                    enable_pass = False
                    break
                    
            if enable_pass:
                pass_path[idx].append(target_idx)
    start = 0
    visited[start] = 0
    cnt = 0
    queue = deque(list(map(lambda x: tuple([0,x]) ,pass_path[0])))
    while queue:
        cnt, next_ = queue.popleft()
        if visited[next_] > cnt+1:
            visited[next_] = cnt+1
            queue += list(map(lambda x: tuple([cnt+1,x]) ,pass_path[next_]))
        else:
            continue
            
    if visited[n-1] == MAX_COUNT:
        print(-1)
    else:
        print(visited[n-1])

solve()
```
